```python
dummy_sex=pd.get_dummies(df_titanic['sex'],drop_first=True)
```

```python
df_titanic_encode_dummies=pd.concat([categorical_feature,dummy_sex],axis=1)
```

```python
df_titanic_encode_dummies.drop('sex',axis=1,inplace=True)
```

```python
df_titanic_encode_dummies['sex_dummies']=df_titanic_encode_dummies['male']
```

```python
df_titanic_encode_dummies.drop('male',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder 

```

```python
ohe=OneHotEncoder()
ohe_df=pd.DataFrame(ohe.fit_transform(df_titanic_encode_dummies[['who']]).toarray(),columns=['child','man','women'])
```

```python
ohe_df
```

```python

df_titanic_encode_dummies=pd.concat([df_titanic_encode_dummies.reset_index(drop=True),ohe_df],axis=1)
df_titanic_encode_dummies.drop('who',axis=1,inplace=True)
```

```python
df_titanic_encode_dummies
```

```python
ohe_fit_town=pd.DataFrame(ohe.fit_transform(df_titanic_encode_dummies[['embark_town']]).toarray(),columns=['Cherbourg','Queenstown','Southampton'])
```

```python
ohe_fit_town
```

```python
df_titanic_encode_dummies.drop('embark_town',axis=1,inplace=True)
```

```python
df_titanic_encode_dummies
```

```python
df_titanic_encode_dummies=pd.concat([df_titanic_encode_dummies.reset_index(drop=True),ohe_fit_town],axis=1)
```

```python
df_titanic_encode_dummies
```

```python
import sklearn
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder

Le=LabelEncoder()
df_titanic_encode_dummies['Class_Encoded']=Le.fit_transform(df_titanic_encode_dummies['class'])
```

```python
df_titanic_encode_dummies.drop('class',axis=1,inplace=True)
```

```python
df_titanic_encode_dummies
```

```python
frequency=df_titanic_encode_dummies['alone'].value_counts(normalize=True)
```
